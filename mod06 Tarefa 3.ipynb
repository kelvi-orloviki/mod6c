#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc['Categoria'] = pd.cut(sinasc['APGAR5'],
                            bins=[-1, 3, 5, 7, 10],
                            labels=['asfixia grave', 'asfixia moderada', 'asfixia leve', 'normal'])
sinasc
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sinasc['SEXO'].value_counts()
```

```python
dfdist_ap_sexo = sinasc[['APGAR5', 'SEXO', 'GESTACAO', 'Categoria']]
dfdist_ap_sexo
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=dfdist_ap_sexo, x="GESTACAO", y="APGAR5", hue="SEXO")

plt.title("Distribuição do APGAR5 por Sexo e Grupo de Tempo de Gestação")
plt.xlabel("Grupo de Tempo de Gestação")
plt.ylabel("APGAR5")
plt.legend(title="Sexo")
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
apgar5xgestacao = pd.crosstab(sinasc['GESTACAO'], sinasc['APGAR5'])
apgar5xgestacao
```

```python
apgar5xgestacao.columns
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
apgar5xgestacao_num = apgar5xgestacao.select_dtypes(include='number')
apgar5xgestacao['Percentual'] = apgar5xgestacao_num.sum(axis=1) / apgar5xgestacao_num.values.sum() * 100
apgar5xgestacao['Percentual'] = apgar5xgestacao['Percentual'].round(2)
apgar5xgestacao
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
apgar5xgestacao_sem_percentual = apgar5xgestacao.drop(columns=["Percentual"])
```

```python
apgar5xgestacao_sem_percentual.plot(
    kind='bar', 
    stacked=True,
    figsize=(10, 6)
)

plt.title("Distribuição de APGAR5 por Idade Gestacional")
plt.xlabel("Idade Gestacional")
plt.ylabel("Frequência")
plt.legend(title="APGAR5")
plt.show()
```

Analisando o gráfico, é possível ver que a partir das 37 semanas a proporção de APGAR5 entre 9 e 10 é muito superior que os demais. A idade gestacional anterior existe uma maior quantidade de APGAR5 entre 9 e 10, mas proporcionalmente a partir de 37 semanas aumenta muito. Isso corrobora com a afirmação dos médicos.
